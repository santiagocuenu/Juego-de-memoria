```python
import random
def crear_tablero():
  simbolos = ["🐶", "🐱", "🐭", "🐹", "🐰", "🦊", "🐻", "🐼", "🐨", "🐯", "🦁", "🐮", "🐷", "🐸", "🐵"]
  tablero = simbolos * 2
  random.shuffle(tablero)
  return tablero
```

```python
def crear_tablero_visible():
 tablero_visible = ["#"] * 30
 return tablero_visible
```

```python
def mostrar_tablero(tablero_visible):
    print("TABLERO ACTUAL ")
    for i in range(6):
      fila = tablero_visible[i*5 : (i+1)*5]
      print("  ".join(fila))
```

```python
def mov_jugador(tablero, tablero_visible):
    print("TU TURNO")
    valido1 = False
    while not valido1:
        try:
            posi1 = int(input("Ingresa la posición de la primera carta (0 al 29): "))
            if posi1 < 0 or posi1 > 29:
                print(" Error: La posición debe estar entre 0 y 29.")
            elif tablero_visible[posi1] != "#":
                print(" Error: Esa carta ya está descubierta. Elige otra.")
            else:
                valido1 = True
        except ValueError:
            print("Error: Por favor ingresa solo números enteros.")
    tablero_visible[posi1] = tablero[posi1]
    mostrar_tablero(tablero_visible)

    valido2 = False
    while not valido2:
        try:
            posi2 = int(input("Ingresa la posición de la segunda carta (0 al 29): "))
            if posi2 < 0 or posi2 > 29:
                print(" Error: La posición debe estar entre 0 y 29.")
            elif tablero_visible[posi2] != "#":
                print("Error: Esa carta ya está descubierta. Elige otra.")
            elif posi1 == posi2:
                print("Error: ¡No puedes elegir la misma carta dos veces!")
            else:
                valido2 = True
        except ValueError:
            print("Error: Por favor ingresa solo números enteros.")
    tablero_visible[posi2] = tablero[posi2]
    mostrar_tablero(tablero_visible)

    if tablero[posi1] == tablero[posi2]:
        print("¡Excelente! Encontraste un par.")
        return 2
    else:
        print("Fallaste. Las cartas se vuelven a voltear.")
        tablero_visible[posi1] = "#"
        tablero_visible[posi2] = "#"
        return -1
```

```python
import random

def mov_maquina(tablero, tablero_visible):
    print("TURNO DE LA MÁQUINA ")
    disponibles = []
    for i in range(30):
        if tablero_visible[i] == "#":
            disponibles.append(i)

    if len(disponibles) < 2:
        print("No hay suficientes cartas disponibles para que la máquina haga un movimiento.")
        return 0

    posi1 = random.choice(disponibles)
    disponibles.remove(posi1)
    posi2 = random.choice(disponibles)

    print(f"La máquina eligió las posiciones {posi1} y {posi2}")
    tablero_visible[posi1] = tablero[posi1]
    tablero_visible[posi2] = tablero[posi2]
    mostrar_tablero(tablero_visible)

    if tablero[posi1] == tablero[posi2]:
        print("¡Oh no! La máquina encontró un par.")
        return 2
    else:
        print("La máquina falló. Las cartas se vuelven a voltear.")
        tablero_visible[posi1] = "#"
        tablero_visible[posi2] = "#"
        return -1
```

```python
def pares_restantes(tablero_visible):
  return "#" in tablero_visible
```

```python
def principal():
    print("¡BIENVENIDO AL JUEGO DE MEMORIA! ")

    jugar_de_nuevo = "si"
    while jugar_de_nuevo.lower() == "si":
        tablero = crear_tablero()
        tablero_visible = crear_tablero_visible()
        puntos_jugador = 20
        puntos_maquina = 20
        mostrar_tablero(tablero_visible)
        while pares_restantes(tablero_visible):
            puntos_jugador += mov_jugador(tablero, tablero_visible)
            print(f"📊 MARCADOR: Tú: {puntos_jugador} pts | Máquina: {puntos_maquina} pts")

            if not pares_restantes(tablero_visible):
                break
            puntos_maquina += mov_maquina(tablero, tablero_visible)
            print(f" MARCADOR: Tú: {puntos_jugador} pts | Máquina: {puntos_maquina} pts")
        print(" " + "="*30)
        print("¡FIN DEL JUEGO! ")
        print("="*30)

        if puntos_jugador > puntos_maquina:
            print(f"¡FELICIDADES! Ganaste con {puntos_jugador} puntos.")
        elif puntos_maquina > puntos_jugador:
            print(f"LA MÁQUINA GANA con {puntos_maquina} puntos. ¡Mejor suerte la próxima!")
        else:
            print("¡ES UN EMPATE IMPRESIONANTE!")

        jugar_de_nuevo = input("\n¿Quieres jugar de nuevo? (si/no): ")
    print("\n¡Gracias por jugar! Hasta la próxima.")
```
